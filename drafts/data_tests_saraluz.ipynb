```python
import matplotlib.pyplot as plt

# Convertir a conjuntos los client_id de ambos DataFrames
clients_inscritos = set(df_exp_clients['client_id'])
clients_en_base_datos = set(df_demo['client_id'])

# 1. Clientes que están inscritos en la prueba y también en la base de datos (intersección)
clients_realized = clients_inscritos.intersection(clients_en_base_datos)

# 2. Clientes que están inscritos en la prueba pero no en la base de datos (diferencia)
clients_not_in_db = clients_inscritos - clients_en_base_datos

# 3. Clientes que están en la base de datos pero no inscritos en la prueba
clients_not_signed_up = clients_en_base_datos - clients_inscritos

# Crear un diccionario con los resultados
result = {
    'Inscritos en la prueba (en base de datos)': len(clients_realized),
    'Inscritos en la prueba (no en base de datos)': len(clients_not_in_db),
    'No inscritos en la prueba (en base de datos)': len(clients_not_signed_up)
}

# Función para formatear las etiquetas mostrando porcentaje y número total
def simple_autopct(pct):
    total = sum(result.values())
    absolute = int(round(pct * total / 100.0))
    return f'{pct:.1f}%\n({absolute:d})'

# Crear la gráfica de tarta
plt.figure(figsize=(7, 7))
plt.pie(result.values(), labels=result.keys(), autopct=simple_autopct, startangle=90, colors=['#66b3ff', '#ff9999', '#99ff99'])
plt.title('Distribución de clientes inscritos y en base de datos')
plt.show()
```

```python
from scipy import stats

# Assuming 'variation' is 1 for experiment and 0 for control
exp_group = df_wd_client_data[df_wd_client_data['Variation'] == "Control"]['bal']
control_group = df_wd_client_data[df_wd_client_data['Variation'] == "Test"]['bal']

# T-test for equality of means
t_stat, p_val = stats.ttest_ind(exp_group, control_group)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

# Interpretation
if p_val < 0.05:
    print("Statistically significant difference between the two groups.")
else:
    print("No statistically significant difference between the two groups.")
```

```python
# Chi-Square Test for Independence
contingency_table = pd.crosstab(df_wd_client_data['gendr'], df_wd_client_data['Variation'])
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")

# Interpretation
if p < 0.05:
    print("There is a significant association between gender and variation.")
else:
    print("No significant association between gender and variation.")
```

```python
# ANOVA to test if balance differs across different age groups
df_wd_client_data['age_group'] = pd.cut(df_wd_client_data['clnt_age'], bins=[0, 25, 35, 45, 55, 65, 100], labels=['<25', '25-35', '35-45', '45-55', '55-65', '>65'])

anova_result = stats.f_oneway(
    df_wd_client_data[df_wd_client_data['age_group'] == '<25']['bal'],
    df_wd_client_data[df_wd_client_data['age_group'] == '25-35']['bal'],
    df_wd_client_data[df_wd_client_data['age_group'] == '35-45']['bal'],
    df_wd_client_data[df_wd_client_data['age_group'] == '45-55']['bal'],
    df_wd_client_data[df_wd_client_data['age_group'] == '55-65']['bal'],
    df_wd_client_data[df_wd_client_data['age_group'] == '>65']['bal']
)

print(f"ANOVA F-Statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")
```

```python
# Definir el orden correcto del proceso
correct_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Función para verificar si los pasos siguen el orden correcto basándose en date_time
def check_order(df_visit):
    # Ordenar el DataFrame por la columna date_time
    df_visit_sorted = df_visit.sort_values('date_time')
    
    # Obtener la lista de process_step en el orden temporal
    process_steps = df_visit_sorted['process_step'].tolist()
    
    # Encontrar la posición de los pasos correctos en la lista actual
    step_positions = [process_steps.index(step) if step in process_steps else -1 for step in correct_order]
    
    # Verificar si el orden está bien (las posiciones deben estar en orden ascendente)
    for i in range(1, len(step_positions)):
        if step_positions[i] != -1 and (step_positions[i] < step_positions[i - 1] or step_positions[i - 1] == -1):
            return False
    
    return True

# Agrupar por visit_id y aplicar la función para verificar si el proceso sigue el orden correcto según date_time
df_invalid_visits = df_time_client.groupby('visit_id').apply(lambda x: not check_order(x))

# Filtrar las visit_id que no siguen el orden correcto
invalid_visits = df_invalid_visits[df_invalid_visits].index

# Obtener los client_id que tienen visitas con errores en el orden de los pasos
df_invalid_clients = df_time_client[df_time_client['visit_id'].isin(invalid_visits)]

# Mostrar las primeras filas con client_id y visit_id que no siguen el orden
df_invalid_clients
```

```python
import pandas as pd

# Definir el orden correcto del proceso con su valor numérico asociado
process_order = {
    'start': 0,
    'step_1': 1,
    'step_2': 2,
    'step_3': 3,
    'confirm': 4
}

# Función para calcular métricas por cliente
def calculate_process_metrics(df_client):
    # Inicializar métricas
    number_try = 0
    process_fails = 0
    process_complete = 0
    time_fails = []
    time_complete = []
    
    # Filtrar todos los intentos del cliente (cada intento empieza con "start")
    starts = df_client[df_client['process_step'] == 'start']
    
    # Recorrer cada intento (cada "start" indica un nuevo intento de proceso)
    for _, start_row in starts.iterrows():
        # Obtener el subset de pasos para este intento en particular
        df_visit = df_client[(df_client['visit_id'] == start_row['visit_id']) & 
                             (df_client['date_time'] >= start_row['date_time'])]

        # Ordenar los pasos por `date_time`
        df_visit_sorted = df_visit.sort_values('date_time')
        
        # Verificar si los pasos siguen el orden permitido (pueden avanzar o retroceder en 1 unidad)
        process_steps = df_visit_sorted['process_step'].tolist()
        process_numbers = [process_order[step] for step in process_steps]
        
        # Comprobar que los pasos avanzan o retroceden como máximo en una unidad
        step_diff = [process_numbers[i+1] - process_numbers[i] for i in range(len(process_numbers)-1)]
        
        # Verificar si todos los cambios de paso están dentro del rango permitido (-1 o 1)
        valid_process = all(diff in [-1, 1] for diff in step_diff)
        
        # Si el proceso es válido y llega a 'confirm', es un proceso completo
        if valid_process and process_steps[-1] == 'confirm':
            process_complete += 1
            start_time = df_visit_sorted['date_time'].iloc[0]
            end_time = df_visit_sorted['date_time'].iloc[-1]
            time_complete.append((end_time - start_time).total_seconds())
        else:
            # Es un fallo
            process_fails += 1
            start_time = df_visit_sorted['date_time'].iloc[0]
            end_time = df_visit_sorted['date_time'].iloc[-1]
            time_fails.append((end_time - start_time).total_seconds())
        
        # Contar como intento
        number_try += 1
    
    # Calcular los tiempos totales
    time_fails = sum(time_fails) if time_fails else 0
    time_complete = sum(time_complete) if time_complete else 0
    
    # Calcular los promedios basados en el número de process_fails y process_complete
    avg_time_fails = round(time_fails / process_fails, 2) if process_fails > 0 else 0
    avg_time_complete = round(time_complete / process_complete, 2) if process_complete > 0 else 0

    # Retornar métricas
    return pd.Series({
        'number_try': number_try,
        'process_fails': process_fails,
        'process_complete': process_complete,
        'time_fails': time_fails,
        'time_complete': time_complete,
        'avg_time_fails': avg_time_fails,
        'avg_time_complete': avg_time_complete
    })

# Agrupar por client_id y aplicar la función de métricas
df_client_metrics = df_time_client.groupby('client_id', group_keys=False).apply(calculate_process_metrics).reset_index()

# Mostrar el resultado
df_client_metrics
```

```python
# Comparación de balances entre los clientes en la variación de experimento Prueba:T-test para comparar los balances (bal) de clientes que fueron parte del experimento (variation=1) frente a los que no lo fueron (variation=0).

from scipy.stats import ttest_ind

# Dividir los datos entre clientes en el experimento y los que no
exp_clients = df_unique_client[df_unique_client['Variation'] == "Control"]['bal']
non_exp_clients = df_unique_client[df_unique_client['Variation'] == "Test"]['bal']

# Realizar el t-test
t_stat, p_value = ttest_ind(exp_clients, non_exp_clients)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

Si el p-value es menor a 0.05, se rechaza la hipótesis nula, lo que indicaría que existe una diferencia estadísticamente significativa entre los balances de los clientes que participaron en el experimento frente a los que no.
Si el p-value es mayor a 0.05, no habría suficiente evidencia para concluir que los balances son diferentes entre los grupos.
